```python
df_covid = pd.read_csv('covidvaccine.csv')
df_covid
```

# Data Cleaning and Pre-processing

```python
### convert text column to lower case letters
df_covid.text = df_covid.text.str.lower()
df_covid.head()
```

```python
### remove punctuation marks

df_covid['text_punc'] = df_covid['text'].str.replace('[^\w\s]','')
df_covid.head()
```

```python
#Removal of stop words

nltk.download('stopwords')
stop = stopwords.words('english')

df_covid['text_punc'] = df_covid['text_punc'].astype(str)

df_covid['text_stop'] = df_covid['text_punc'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df_covid.head()
```

```python
#Tokenization of Tweets

def tokenization(text):
    text = re.split('\W+', text)
    return text

df_covid['text_tokenized'] = df_covid['text_stop'].apply(lambda x: tokenization(x.lower()))
df_covid[['date','text', 'text_punc', 'text_stop', 'text_tokenized']][0:9]
```

```python
nltk.download('omw-1.4')
```

```python
### Lemmenisation step

nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

df_covid['text_lemmatized'] = df_covid['text_tokenized'].apply(lambda x: lemmatizer(x))
df_covid[['date','text', 'text_punc', 'text_tokenized','text_stop','text_lemmatized']][0:9]
```

```python
### removing unnecessary columns from dataframe
df_covid = df_covid.drop(columns= ['user_name','user_description', 'user_friends', 'user_favourites', 'source', 'is_retweet', 'text', 'text_punc', 'text_tokenized', 'text_stop', 'text_tokenized'])

df_covid
```
