 Having collected 10,000 tweets from December 2020, again the bottleneck I encountered was that the tweets were collected chronologically and not at random across December and so I decided to no longer pursue this method for collecting tweets.


Below it the final version of my web-scraping code whereby I collect 1000 tweets per day, with the tweets collected being stored in a csv which I will then import into a dataframe for further use.
This method is more successful in providing a more representative dataset and will be used for all further processing and sentiment analysis 

```python
from tqdm.auto import tqdm
from time import sleep
from datetime import date
from datetime import timedelta
import os
import glob
```

```python
def GrabTweets(month,year,num_tweets_per_day,query):

    # Calculate Total Tweets Per Month to Collect, given some months have fewer days than some I set limits to represent this
    if month in [1, 3, 5, 7, 8, 10, 12]:
        days_in_month = 31
        total_tweets = days_in_month * num_tweets_per_day

    elif month in [4, 6, 9, 11]:
        days_in_month = 30
        total_tweets = days_in_month * num_tweets_per_day

    elif month == 2:
        days_in_month = 28
        total_tweets = days_in_month * num_tweets_per_day
        

    # Creating List to Append Tweet Data to
    tweets_list = []
    pbar = tqdm(total=total_tweets)

    # Create Date Object
    since = date(year, month, 1).isoformat()
    until = date(year, month, 2).isoformat()

    # Counter
    counter = 0
    DayCounter = 0

    # Using TwitterSearchScraper to Scrape Data on Covid-19 vaccine in X Month in 2021 & Append Tweets to List
    while DayCounter < days_in_month:
        for i, tweet in enumerate(sntwitter.TwitterSearchScraper('{} since:{} until:{}'.format(query, since, until)).get_items()):
            if i >= num_tweets_per_day:
                    until = (date.fromisoformat(until)+timedelta(days=1)).isoformat()
                    break
            tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.replyCount, tweet.retweetCount, tweet.likeCount, tweet.quoteCount])
            sleep(0.01)
            counter += 1
            pbar.update(1)
        
        since = (date.fromisoformat(since) + timedelta(days=1)).isoformat()
        until = (date.fromisoformat(until) + timedelta(days=1)).isoformat()
        counter += 1
        DayCounter += 1

    pbar.close()

    # Creating a DataFrame from the Tweets List Above
    DF = pd.DataFrame(tweets_list, columns=['Date', 'Tweet ID', 'Text', 'Username', 'Reply Count', 'Retweet Count', 'Like Count', 'Quote Count'])

    
    # Write to CSV
    DF.to_csv('{}-Tweets-{}-{}-{}.csv'.format(query,month,year,total_tweets), encoding='utf-8', index=False)
    
    # Print Done
    print(pbar,"\n")
    print("Done!")
    
    
```

```python
### Collecting 1000 tweets each day of 2021

query = "covid vaccine"
year = 2021
num_tweets_per_day = 1000

### Run each iteration individually to avoid over-working the computer
### do this by placing a hashtag before any unwanted lines
GrabTweets(1,year,num_tweets_per_day,query)
GrabTweets(2,year,num_tweets_per_day,query)
GrabTweets(3,year,num_tweets_per_day,query)
GrabTweets(4,year,num_tweets_per_day,query)
GrabTweets(5,year,num_tweets_per_day,query)
GrabTweets(6,year,num_tweets_per_day,query)
GrabTweets(7,year,num_tweets_per_day,query)
GrabTweets(8,year,num_tweets_per_day,query)
GrabTweets(9,year,num_tweets_per_day,query)
GrabTweets(10,year,num_tweets_per_day,query)
GrabTweets(11,year,num_tweets_per_day,query)
GrabTweets(12,year,num_tweets_per_day,query)

    
```

# Data Cleaning and Pre-processing

```python
### Cleaning 2021 Tweets
Tweets_2021.Text = Tweets_2021.Text.str.lower()

### Removing punctuation marks
Tweets_2021['text_punc'] = Tweets_2021['Text'].str.replace('[^\w\s]','')

### remove stopwords
stop = stopwords.words('english')

stop.append("covid vaccine")
stop.append("covid19 vaccine")

Tweets_2021['text_punc'] = Tweets_2021['text_punc'].astype(str)

Tweets_2021['text_stop'] = Tweets_2021['text_punc'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
Tweets_2021.head()
```

```python
### Tokenisation of Tweets

def tokenization(text):
    text = re.split('\W+', text)
    return text
Tweets_2021['text_tokenized'] = Tweets_2021['text_stop'].apply(lambda x: tokenization(x.lower()))
Tweets_2021[['Date','Text', 'text_punc', 'text_stop', 'text_tokenized']][0:9]
```

```python
### Lemminisation of tweets

wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

Tweets_2021['text_lemmatized'] = Tweets_2021['text_tokenized'].apply(lambda x: lemmatizer(x))
Tweets_2021[['Date', 'Text', 'text_punc', 'text_tokenized','text_stop','text_lemmatized']][0:9]
```

```python
### removing unnecessary columns from dataframe
Tweets_2021 = Tweets_2021.drop(columns= ['Text', 'text_punc', 'text_tokenized', 'text_stop', 'Tweet ID', 'Username', 'Reply Count', 'Retweet Count', 'Like Count', 'Quote Count'])

Tweets_2021
```
