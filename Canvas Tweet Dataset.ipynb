```python
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('omw-1.4')
from nltk.corpus import stopwords
```

The third tweet library I used to explore sentiment of Tweets discussing the Covid-19 Vaccine was the dataset provided on 
Canvas. The dataset is labelled vaccination_tweets

```python
### using vaccination_tweets dataset provided to students on Canvas
covid_tweets = pd.read_csv('vaccination_tweets.csv')
covid_tweets
```

# Data Cleaning and Pre-processing

```python
### setting text to lower case
covid_tweets.text = covid_tweets.text.str.lower()

### Removing punctuation marks
covid_tweets['text_punc'] = covid_tweets['text'].str.replace('[^\w\s]','')

### remove stopwords
nltk.download('stopwords')
stop = stopwords.words('english')

### adding new columns to the dataframe
covid_tweets['text_punc'] = covid_tweets['text_punc'].astype(str)
covid_tweets['text_stop'] = covid_tweets['text_punc'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
covid_tweets
```

```python
### Tokenisation of Tweets

def tokenization(text):
    text = re.split('\W+', text)
    return text

### add a text tokenised column to the dataset
covid_tweets['text_tokenized'] = covid_tweets['text_stop'].apply(lambda x: tokenization(x.lower()))
covid_tweets[['date','text', 'text_punc', 'text_stop', 'text_tokenized']][0:9]
```

```python
### Text lemmatisation

nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

covid_tweets['text_lemmatized'] = covid_tweets['text_tokenized'].apply(lambda x: lemmatizer(x))
covid_tweets[['date', 'text', 'text_punc', 'text_tokenized','text_stop','text_lemmatized']][0:9]
```

```python
covid_tweets
```

```python
### Removing unnecessary columns from dataframe
covid_tweets = covid_tweets.drop(columns = ['id', 'user_name', 'user_description', 'user_friends', 'user_favourites', 'text', 'source',
                                            'retweets', 'favorites', 'is_retweet', 'text_punc', 'text_stop', 'text_tokenized'])
covid_tweets
```

```python
### Writing cleaned dataframe to new csv for storage
covid_tweets.to_csv('Tweets_Canvas_2021_Cleaned.csv')
```
